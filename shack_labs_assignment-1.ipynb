```python
# LINEAR REGRESSION

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics

X = df[['House Age','Distance from nearest Metro station (km)','Number of convenience stores','latitude','longitude','Number of bedrooms','House size (sqft)']]
y = df['House price of unit area']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)

lm = LinearRegression()
lm.fit(X_train,y_train)

print(lm.intercept_)
lm.coef_

cdf = pd.DataFrame(lm.coef_,X.columns,columns=['Coeff'])
cdf

predictions = lm.predict(X_test)
plt.scatter(y_test,predictions)

sns.distplot((y_test-predictions),bins=50);

print("The accuracy of the model is",lm.score(X_test,y_test)*100)

metrics.mean_absolute_error(y_test,predictions)
metrics.mean_squared_error(y_test,predictions)
np.sqrt(metrics.mean_squared_error(y_test,predictions))
```

```python
# DECISION TREE REGRESSOR

from sklearn.tree import DecisionTreeRegressor

dtree = DecisionTreeRegressor()
dtree.fit(X_train,y_train)

predictions = dtree.predict(X_test)
plt.scatter(y_test,predictions)

print("The accuracy of the model is",dtree.score(X_test,y_test)*100)
```

```python
# LASSO REGRESSION

from sklearn.linear_model import Lasso

lasso = Lasso()
lasso.fit(X_train,y_train)

predictions = lasso.predict(X_test)
plt.scatter(y_test,predictions)

print("The accuracy of the model is",lasso.score(X_test,y_test)*100)
```
